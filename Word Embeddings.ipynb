### Word to Integer

```python
def word_to_integer(vocab, word):
    if word in vocab:
        integer = vocab.index(word) + 1
    else:
        integer = 0
        
    return integer
        
# Test

print(word_to_integer(vocab, 'was'))
print(word_to_integer(vocab, 'asfqwdgdfsf'))
```

```python
def integer_to_word(vocab, integer):
    if 0 < integer < len(vocab):
        word = vocab[integer-1]   
    elif integer == 0:
        word = '<unk>'   
    elif integer >= len(vocab):
        word = None
        print("Index is too high")
    
    return word

# Test

print(integer_to_word(vocab, 22))
print(integer_to_word(vocab, 0))
print(integer_to_word(vocab, 1000000000))

        
```

### Integer to One-Hot Vector

```python
def int_to_vec(integer, vocab_len):
    l = [0 for i in range(vocab_len+1)]
    l[integer] = 1
    vector = np.asarray(l)
    vector = vector.reshape(-1, 1)
    return vector

# Test

vec = int_to_vec(1, 7)
print(vec)
print(vec.shape)
```

### Creating Windows

```python
def create_windows(window_size, corpus):
    windows = []
    for sentence in corpus:
        for i in range(window_size, len(sentence)-window_size):
            window = sentence[i-window_size:i+window_size+1]
            windows.append(window)
    return windows
            
windows = create_windows(2, tokenized_Corpus)

# Test

#print(windows[:100])

# Or

for window in windows[:50]:
    words = []
    for integer in window:
        word = integer_to_word(vocab, integer)
        words.append(word)
        
    print(words)
    #print(len(words))
    
        
    
```

### Generating Context and Target Vectors

```python
def generate_context_target(window, window_size, vocab_len):
    target_word = window[window_size]
    context_words = window[:window_size] + window[window_size+1:]
    
    target_vector = int_to_vec(target_word, vocab_len)
    context_vectors = []
    for word in context_words:
        context_vectors.append(int_to_vec(word, vocab_len))
    
    return target_vector, context_vectors
    # NOTE: context_vectors is a list of vectors

# Test

window = ["Hi", "my", "name", "is", "Valensus"]
window = [word_to_integer(vocab, word) for word in window]
window_size = 2

target_vector_test, context_vectors_test = generate_context_target(window, window_size, len(vocab))
print(target_vector_test, '\n\n', context_vectors_test)
```

### Averaging Context Vectors

```python
def average_context_vectors(context_vectors):
    context_vectors_added = sum(context_vectors)
    context_vectors_average = context_vectors_added /len(context_vectors)
    
    return context_vectors_average

# Test

print(average_context_vectors(context_vectors_test))
```

### Initialize Model

```python
def initialize_model(embedding_length, vocab_length):
    W1 = np.random.rand(embedding_length, vocab_length+1)
    W2 = np.random.rand(vocab_length+1, embedding_length)
    b1 = np.random.rand(embedding_length, 1)
    b2 = np.random.rand(vocab_length+1, 1)
    
    return W1, W2, b1, b2
    
```

### Forward Propagation

```python
def forward_prop(x, W1, W2, b1, b2):
    
    z1 = np.dot(W1, x) + b1
    z1 = np.maximum(0,z1)
    z2 = np.dot(W2, z1) + b2
    # Softmax
    e_z = np.exp(z2)
    yhat = e_z/np.sum(e_z,axis=0)
    
    return yhat, z2, z1
```

### Cost Function

```python
def compute_cost(y, yhat, batch_size):
    # cost function 
    logprobs = np.multiply(np.log(yhat),y) + np.multiply(np.log(1 - yhat), 1 - y)
    cost = - 1/batch_size * np.sum(logprobs)
    cost = np.squeeze(cost)
    return cost
```

### Backpropagation

```python
def back_prop(x, yhat, y, z1, W1, W2, b1, b2, batch_size):
    
    l1 = np.dot(W2.T,(yhat-y))

    l1 = np.maximum(0,l1)

    grad_W1 = (1/batch_size)*np.dot(l1,x.T) 

    grad_W2 = (1/batch_size)*np.dot(yhat-y,z1.T)

    grad_b1 = np.sum((1/batch_size)*np.dot(l1,x.T),axis=1,keepdims=True)

    grad_b2 = np.sum((1/batch_size)*np.dot(yhat-y,z1.T),axis=1,keepdims=True)
    
    return grad_W1, grad_W2, grad_b1, grad_b2

    
```

### Updating Weights and Biases

```python
def update_parameters(W1, W2, b1, b2, grad_W1, grad_W2, grad_b1, grad_b2, alpha):
    
    W1 -= alpha*grad_W1 
    W2 -= alpha*grad_W2
    b1 -= alpha*grad_b1
    b2 -= alpha*grad_b2
    
    return W1, W2, b1, b2
    
    
```

### Creating Batches 

```python
def create_batches(inputs, batch_size):
    
    batches = []
    stop = len(inputs)//batch_size+2
    index = 0
    for i in range(1, stop):
        batch = inputs[index:index+batch_size]
        index += batch_size
        batches.append(batch)
    
    return batches

Batches = create_batches(windows, 64)
for i in range(1):
    print(Batches[1532:])
    print(len(Batches[i]))
    print(len(Batches))
    print(len(windows)//64+1)
    
    
```

### Batch to Inputs and Targets

```python
def process_batch(batch, window_size, vocab_len):
    
    batch_processed_X = []
    batch_processed_Y = []
    
    for window in batch:
        
        target_vector, context_vectors = generate_context_target(window, window_size, vocab_len)
        
        context_vector = average_context_vectors(context_vectors)
        
        batch_processed_X.append(context_vector)
        batch_processed_Y.append(target_vector)
    
    X = np.stack(batch_processed_X, axis = 1)
    X = X.reshape(X.shape[0],X.shape[1])
    Y = np.stack(batch_processed_Y, axis = 1)
    Y = Y.reshape(Y.shape[0],Y.shape[1])
    
    
    return X, Y
        
# TEST

X_test, Y_test = process_batch(Batches[1], window_size, len(vocab))
    
print(X_test)
print(X_test.shape)
print(Y_test.shape)
print(len(vocab))
```

### Create the Model

```python
def model(Batches, embedding_length, vocab_length, epochs, window_size, alpha=0.03, verbose = False):
    
    W1, W2, b1, b2 = initialize_model(embedding_length, vocab_length)
    counter = 0
    
    for i in range(epochs):
        for batch in Batches:
            batch_size = len(batch)
            
            X, Y = process_batch(batch, window_size, vocab_length)
            
            yhat, _, z1 = forward_prop(X, W1, W2, b1, b2)
            
            
            
            cost = compute_cost(Y, yhat, batch_size)
            
            grad_W1, grad_W2, grad_b1, grad_b2 = back_prop(X, yhat, Y, z1, W1, W2, b1, b2, batch_size)
        
            W1, W2, b1, b2 = update_parameters(W1, W2, b1, b2, grad_W1, grad_W2, grad_b1, grad_b2, alpha)
            
            if verbose:
                counter += 1
                if counter % 100 == 0:
                    print('Cost: {}'.format(cost))
                    print('\n')
                
        
    
    return W1, W2, b1, b2
        
    
    
    
    
```

```python
model(Batches, 2, len(vocab), 1, 2, alpha=0.03, verbose = True)
```
